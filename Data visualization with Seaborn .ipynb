```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
from numpy import mean, median, max, min 
```

```python
sns.barplot(x='smoker', y ='total_bill', data=tips, hue = "sex", palette = "spring", order = ['No', 'Yes'], estimator = max)
```

```python
sns.barplot(x='smoker', y ='total_bill', data=tips, hue = "sex", palette = "spring", order = ['No', 'Yes'], estimator = mean)
```

```python
sns.barplot(x='smoker', y ='total_bill', data=tips, hue = "sex", palette = "husl", order = ['No', 'Yes'], estimator = median
                      , capsize = .05, ci= 99.999)
# color = 'red, saturation = .5
```

```python
dist = sns.distplot(tips['total_bill'], color = 'blue')
```
